```python
import cv2
import pytesseract
from transformers import pipeline
import re

# Initialize the summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Path to the video file
video_path = "/content/drive/MyDrive/imagesface/video/WhatsApp Video 2024-03-28 at 21.12.01_2c72c7d1.mp4"

# Open the video file
cap = cv2.VideoCapture(video_path)

# Check if the video file opened successfully
if not cap.isOpened():
    print("Error: Unable to open video file.")
    exit()

# Variable to store the text of the previous frame
prev_frame_text = ""

# Loop through each frame in the video
while cap.isOpened():
    # Read the frame
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Perform OCR to extract text from the frame
    extracted_text = pytesseract.image_to_string(gray_frame)

    # Remove symbols from the extracted text
    extracted_text_cleaned = re.sub(r'[^\w\s]', '', extracted_text)

    # Check if the extracted text is the same as the text from the previous frame
    if extracted_text_cleaned == prev_frame_text:
        continue  # Skip summarization and printing for this frame

    # Summarize the extracted text
    summary = summarizer(extracted_text_cleaned, max_length=130, min_length=30, do_sample=False)

    # Extract and print the summary sentence
    summary_sentence = summary[0]['summary_text']
    print(summary_sentence)
    #print("\n")

    # Update the previous frame text
    prev_frame_text = extracted_text_cleaned

    # Press 'q' to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close any open windows
cap.release()
cv2.destroyAllWindows()
```
