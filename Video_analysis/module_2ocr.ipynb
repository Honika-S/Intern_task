```python
import cv2
import pytesseract
from datetime import datetime
import string

# Set the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

# Function to extract text from video frame
def extract_text_from_frame(frame):
    try:
        # Convert the frame to grayscale
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        # Perform OCR on the frame
        text = pytesseract.image_to_string(gray_frame)
        return text.strip()
    except Exception as e:
        print(f"Error processing frame: {e}")
        return None

# Function to filter out non-printable characters and meaningless text
def filter_text(text):
    printable = set(string.printable)
    filtered_text = ''.join(filter(lambda x: x in printable, text))  # Remove non-printable characters
    filtered_text = ' '.join(word for word in filtered_text.split() if len(word) > 1)  # Remove single characters
    return filtered_text

# Path to the video file
video_path = '/content/drive/MyDrive/imagesface/video/WhatsApp Video 2024-03-28 at 21.12.01_2c72c7d1.mp4'

# Open the video file
cap = cv2.VideoCapture(video_path)

# Check if the video file opened successfully
if not cap.isOpened():
    print("Error: Unable to open video file.")
    exit()

# Set to store unique extracted text
unique_text_set = set()

# Loop through each frame in the video
while cap.isOpened():
    # Read the frame
    ret, frame = cap.read()
    if not ret:
        break

    # Extract text from the frame
    extracted_text = extract_text_from_frame(frame)
    if extracted_text:
        # Filter text
        filtered_text = filter_text(extracted_text)
        if filtered_text and filtered_text not in unique_text_set:
            # Get the timestamp
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            # Print extracted text and timestamp
            print(f"Timestamp: {timestamp}")
            print("Extracted Text:")
            print(filtered_text)
            # Add extracted text to the set
            unique_text_set.add(filtered_text)

# Release the video capture object and close any open windows
cap.release()
cv2.destroyAllWindows()
```
